```python
#Formatting dates
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'],utc = True)
df['Formatted Date']
```

```python
#Set date as Index
df = df.set_index('Formatted Date')
```

```python
df.head()
```

```python
#after reshampling

df_columns = ['Apparent Temperature (C)','Humidity']
```

```python
df_monthly_sampling = df[df_columns].resample('M').mean()
```

```python
df_monthly_sampling.head()
```

```python
sns.lineplot(x='Apparent Temperature (C)',y='Humidity',data=df_monthly_sampling)
```

```python
#Plotting the variation in Apparent temperature and humidity  with time

plt.figure(figsize=(20,2))
df_monthly_sampling.plot()
plt.title('Variation in Apparent Temperature and humidity with time')
```

From the insight abobe we can say that humidity  remained almost constant in these 10 years .
Even the average apparent temperature is akmost same(as peak lie on the same line) )

```python
#Retriving the data of a particular  month from every year ,
df_april = df_monthly_sampling[df_monthly_sampling.index.month==4]
```
